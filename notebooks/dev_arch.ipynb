```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-2,2, 100)
for k in [0.01, 0.1, 1, 10, 100]:
    Y = np.tanh(k*X)
    plt.plot(X, Y, label=f"k={k}")
plt.legend()
plt.show()

# --> k=100 for very sharp transition
```

```python
X = np.linspace(-100000, 100000, 100)


for sigma in [0.1, 1, 10, 100, 1000, 10000, 100000, np.std(X)]:
    Y = np.exp(-X**2 / sigma ** 2)
    plt.plot(X, Y, label=f"$\\sigma={sigma}$")
    plt.axhline(1, color="k")
plt.legend()
plt.show()

# --> sigma = var(R)
```
