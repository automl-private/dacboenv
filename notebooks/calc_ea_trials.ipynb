```python
import numpy as np
import matplotlib.pyplot as plt
n_workers = np.arange(1, 11)

dim = 8
t_budget = 24 * 60  # min
t_ep = 5  # min

n_pop = dim * 3
t_gen = (t_ep * n_pop * (1/n_workers))
n_gen = t_budget / t_gen
n_trials = n_gen * n_pop
plt.plot(n_workers, n_trials)
print(n_workers)
print(n_trials)
```
